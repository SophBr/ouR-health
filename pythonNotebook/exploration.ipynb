```python
resList = []
topic = 'ADDEPEV2'
for state in df['X_STATE'].unique():
    tmpDf = df[(df['X_STATE'] == state)]
    ratioAr = tmpDf.groupby(topic).count()['IDATE'].values
    if len(ratioAr) == 0:
        continue
    rat = (ratioAr[0])/(ratioAr[0]+ratioAr[1])
    resList.append([rat,state])
resList.sort()
resList
```
